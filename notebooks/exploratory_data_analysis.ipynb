```python
import pandas as pd
from IPython.display import display

# Path to your Excel file
file_path = '../data/raw/ODS Osiris basistabellen Inschrijving etc.xlsx'

# Load the Excel file, excluding the 'vooraf' sheet
excel_data = pd.ExcelFile(file_path)
sheet_names = excel_data.sheet_names

# Remove 'vooraf' from the list of sheet names
if 'vooraf' in sheet_names:
    sheet_names.remove('vooraf')

# Load each sheet into a DataFrame, using the sheet name as the key
dataframes = {}
for sheet_name in sheet_names:
    dataframes[sheet_name] = pd.read_excel(file_path, sheet_name=sheet_name)
    print(f"Loaded {sheet_name} with {len(dataframes[sheet_name])} rows.")
    # Display the DataFrame using IPython's display function
    print(f"Displaying first few rows of {sheet_name}:")
    display(dataframes[sheet_name].head())

# Optionally, you can display a specific DataFrame in a nicer format whenever needed
# Example: Accessing and displaying the 'ODS.OSS_STUDENT_INSCHRIJFHIST' DataFrame
print("Detailed view of 'ODS.OSS_STUDENT_INSCHRIJFHIST':")
display(dataframes['ODS.OSS_STUDENT_INSCHRIJFHIST'].head())
```

```python
# Analyzing each DataFrame
for sheet_name, df in dataframes.items():
    print(f"\nAnalysis for sheet: {sheet_name}")
    # Display basic info about the DataFrame
    print("Basic Info:")
    df.info()
    # Display the number of missing values per column
    print("\nMissing Values:")
    display(df.isna().sum())
    # Display basic statistics for numeric columns
    print("\nStatistics for Numeric Columns:")
    display(df.describe())
    # Display the first few rows to get a glimpse of the data
    print("\nPreview of Data:")
    display(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetics for seaborn plots
sns.set(style="whitegrid")

# Path to your Excel file
file_path = '../data/raw/ODS Osiris basistabellen Inschrijving etc.xlsx'

# Load the Excel file, excluding the 'vooraf' sheet
excel_data = pd.ExcelFile(file_path)
sheet_names = excel_data.sheet_names

# Remove 'vooraf' from the list of sheet names
if 'vooraf' in sheet_names:
    sheet_names.remove('vooraf')

# Load each sheet into a DataFrame, using the sheet name as the key
dataframes = {}
for sheet_name in sheet_names:
    df = pd.read_excel(file_path, sheet_name=sheet_name)
    dataframes[sheet_name] = df
    print(f"Loaded {sheet_name} with {df.shape[0]} rows and {df.shape[1]} columns.")

# Analyzing each DataFrame graphically and numerically
for sheet_name, df in dataframes.items():
    print(f"\nAnalysis for sheet: {sheet_name}")
    
    # Plot missing values heatmap
    plt.figure(figsize=(10, 6))
    sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
    plt.title(f'Missing Data Heatmap for {sheet_name}')
    plt.show()

    # Plot histograms for all numeric columns
    numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
    df[numeric_cols].hist(bins=15, figsize=(15, 6), layout=(2, 4), color='skyblue')
    plt.suptitle(f'Histograms of Numeric Columns in {sheet_name}')
    plt.show()

    # Show correlations among numeric columns if there are multiple numeric columns
    if len(numeric_cols) > 1:
        plt.figure(figsize=(8, 6))
        sns.heatmap(df[numeric_cols].corr(), annot=True, fmt=".2f", cmap='coolwarm')
        plt.title(f'Correlation Matrix for {sheet_name}')
        plt.show()
```
