```python
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer

# Example DataFrame
data = {
    'STUDENTNUMBER': [1, 2, 3],
    'EXAMENTYPE': [['A', 'B'], ['A', 'C'], ['B', 'D']]
}
osiris = pd.DataFrame(data)

# Initialize MultiLabelBinarizer
mlb = MultiLabelBinarizer()

# Apply MultiLabelBinarizer to the column with lists of strings
encoded_columns = mlb.fit_transform(osiris['EXAMENTYPE'])

# Create a DataFrame with the encoded columns
encoded_df = pd.DataFrame(encoded_columns, columns=mlb.classes_)

# Concatenate the encoded columns with the original DataFrame
osiris_encoded = pd.concat([osiris, encoded_df], axis=1).drop('EXAMENTYPE', axis=1)

print(osiris_encoded)
```
