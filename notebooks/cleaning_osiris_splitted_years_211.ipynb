```python
# reshuffle dataframe
osiris = osiris.sample(frac=1).reset_index(drop=True)
```

```python
from sklearn.impute import KNNImputer

# Identify columns with float data type
float_columns = osiris.select_dtypes(include=['float']).columns

# Filter columns that have missing values
float_columns_with_missing_values = [col for col in float_columns if osiris[col].isnull().any()]

print("Float columns with missing values:", float_columns_with_missing_values)

# Create a KNNImputer instance with K=25
imputer = KNNImputer(n_neighbors=25)

# Impute the missing values for the float columns
osiris[float_columns_with_missing_values] = imputer.fit_transform(osiris[float_columns_with_missing_values])

# Verify that there are no missing values left in the float columns
print(osiris[float_columns_with_missing_values].isnull().sum())
```

```python
# Identify columns with object data type
object_columns = osiris.select_dtypes(include=['object']).columns

# Filter columns that have missing values
object_columns_with_missing_values = [col for col in object_columns if osiris[col].isnull().any()]

print("Object columns with missing values:", object_columns_with_missing_values)

# Impute missing values with the mode for each column
for col in object_columns_with_missing_values:
    mode_value = osiris[col].mode()[0]  # Get the mode (most frequent value)
    osiris[col].fillna(mode_value, inplace=True)

# Verify that there are no missing values left in the object columns
print(osiris[object_columns_with_missing_values].isnull().sum())
```

```python
# filter students who have four successive years in the DATE_ENROLLMENT_REQUEST column 
# without any gaps and create separate dataFrames for first-year, second-year, third-year, 
# and fourth-year students based on four consecutive years. 
# This is to do the analysis separately year-wise
```

```python
# Convert DATE_ENROLLMENT_REQUEST to datetime format
osiris['DATE_ENROLLMENT_REQUEST'] = pd.to_datetime(osiris['DATE_ENROLLMENT_REQUEST'], errors='coerce')

# Extract year from DATE_ENROLLMENT_REQUEST
osiris['YEAR'] = osiris['DATE_ENROLLMENT_REQUEST'].dt.year

# Function to check for 4 successive years and return a dictionary with year labels
def label_four_consecutive_years(years):
    years = sorted(years)
    for i in range(len(years) - 3):
        if years[i+3] - years[i] == 3:
            return {years[i]: 'First year', years[i+1]: 'Second year', years[i+2]: 'Third year', years[i+3]: 'Fourth year'}
    return {}

# Apply the labeling function to each student
osiris['YEAR_LABEL'] = None  # Create a new column for year labels

# Loop through each student
for student_num, group in osiris.groupby('STUDENT_NUM'):
    # Get the unique years for this student
    unique_years = group['YEAR'].unique()
    # Label the years if the student has four consecutive years
    year_labels = label_four_consecutive_years(unique_years)
    # Assign the labels to the respective rows
    osiris.loc[osiris['STUDENT_NUM'] == student_num, 'YEAR_LABEL'] = osiris['YEAR'].map(year_labels)

# Now filter for each year
first_year_data = osiris[osiris['YEAR_LABEL'] == 'First year'].drop(columns='YEAR_LABEL')
second_year_data = osiris[osiris['YEAR_LABEL'] == 'Second year'].drop(columns='YEAR_LABEL')
third_year_data = osiris[osiris['YEAR_LABEL'] == 'Third year'].drop(columns='YEAR_LABEL')
fourth_year_data = osiris[osiris['YEAR_LABEL'] == 'Fourth year'].drop(columns='YEAR_LABEL')
```
