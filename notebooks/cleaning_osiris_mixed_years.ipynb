```python
# Define the list of valid qualifications
valid_qualifications = ['VWO', 'HAVO', 'HBO', 'MBO', 'VO', 'WO']

# Function to check and replace values
def replace_qualification_type(row):
    if row['PRIOR_QUALIFICATION_TYPE'] == 'OVERIG':
        # Extract the qualification from the PRIOR_QUALIFICATION column
        qualification = row['PRIOR_QUALIFICATION']
        # Check if the qualification contains one of the valid qualifications
        for q in valid_qualifications:
            if q in qualification:
                return q  # Replace with the valid qualification
    return row['PRIOR_QUALIFICATION_TYPE']  # Keep the original value

# Apply the function to each row
df4['PRIOR_QUALIFICATION_TYPE'] = df4.apply(replace_qualification_type, axis=1)

# Remove any other strings that are not in the valid qualifications list
df4['PRIOR_QUALIFICATION_TYPE'] = df4['PRIOR_QUALIFICATION_TYPE'].apply(lambda x: x if x in valid_qualifications else None)

# Display the updated DataFrame
df4
```

```python
# Missing values imputation: we see there are missing values for float and object data types. 
# For the float, KNN can can be used, its a good practice to reshuffle the dataframe before applying KNN
# For the object, mode can be used 
```

```python
# reshuffle dataframe
osiris = osiris.sample(frac=1).reset_index(drop=True)
```

```python
from sklearn.impute import KNNImputer

# Identify columns with float data type
float_columns = osiris.select_dtypes(include=['float']).columns

# Filter columns that have missing values
float_columns_with_missing_values = [col for col in float_columns if osiris[col].isnull().any()]

print("Float columns with missing values:", float_columns_with_missing_values)

# Create a KNNImputer instance with K=25
imputer = KNNImputer(n_neighbors=25)

# Impute the missing values for the float columns
osiris[float_columns_with_missing_values] = imputer.fit_transform(osiris[float_columns_with_missing_values])

# Verify that there are no missing values left in the float columns
print(osiris[float_columns_with_missing_values].isnull().sum())
```

```python
# Identify columns with object data type
object_columns = osiris.select_dtypes(include=['object']).columns

# Filter columns that have missing values
object_columns_with_missing_values = [col for col in object_columns if osiris[col].isnull().any()]

print("Object columns with missing values:", object_columns_with_missing_values)

# Impute missing values with the mode for each column
for col in object_columns_with_missing_values:
    mode_value = osiris[col].mode()[0]  # Get the mode (most frequent value)
    osiris[col].fillna(mode_value, inplace=True)

# Verify that there are no missing values left in the object columns
print(osiris[object_columns_with_missing_values].isnull().sum())
```

```python
# Function to fill datetime columns with median date
def fill_datetime_columns_with_median(osiris):
    for col in osiris.select_dtypes(include=['datetime']):
        median_date = osiris[col].dropna().median()
        osiris[col].fillna(median_date, inplace=True)
    return osiris

# Apply the function to the DataFrame
osiris = fill_datetime_columns_with_median(osiris)
```

```python
# Define the aggregation functions to use ' '.join for all columns
aggregation_functions = {col: ' '.join for col in osiris.columns if col != 'STUDENT_NUM'}

# Apply groupby and aggregation using ' '.join for all columns
grouped_osiris = osiris.groupby('STUDENT_NUM').agg(aggregation_functions).reset_index()

# Display the grouped DataFrame
print("\nGrouped DataFrame:")
grouped_osiris
```

```python
# keeping only the 'first' string if all strings in a column for a particular STUDENT_NUM are the same, and 
# concatenating the strings if they are different

# Define the aggregation functions to use ' '.join for all columns
aggregation_functions = {col: ' '.join for col in osiris.columns if col != 'STUDENT_NUM'}

# Apply groupby and aggregation using ' '.join for all columns
grouped_osiris = osiris.groupby('STUDENT_NUM').agg(aggregation_functions).reset_index()

# Function to remove duplicate strings and keep only one occurrence
def remove_duplicates(text):
    words = text.split()
    unique_words = list(dict.fromkeys(words))  # Keep only the first occurrence
    return ' '.join(unique_words)

# Apply the remove_duplicates function to all columns except 'STUDENT_NUM'
for col in grouped_osiris.columns:
    if col != 'STUDENT_NUM':
        grouped_osiris[col] = grouped_osiris[col].apply(remove_duplicates)

# Display the cleaned DataFrame
print("\nCleaned Grouped DataFrame with Unique Strings:")
grouped_osiris
```
